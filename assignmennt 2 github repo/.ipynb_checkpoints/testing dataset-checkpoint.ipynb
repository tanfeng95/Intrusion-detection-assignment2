```python
# Prepare dataset
dataset_log = pd.read_csv('./capture 2 training splitted.csv' ,sep= ',')
dataset_log.head()
```

```python
df =  dataset_log[['Address A','Port A','Address B','Port B','Packets','Bytes',
                    'Stream ID','Total Packets','Percent Filtered','Packets A to B',
                    'Bytes A to B','Packets B to A','Bytes B to A','Rel Start',
                    'Duration','Bits A to B','Bits B to A','classification']]
df.columns
```

```python
df.dropna(subset=['Bits B to A'], inplace=True)
df.dropna(subset=['Bits A to B'], inplace=True)
df.dropna(subset=['classification'], inplace=True)
df.dropna(subset=['Bytes B to A'], inplace=True)
df
# remove the blanks 
```

```python
# Iterate over each row
for index, row in df.iterrows():
    # Check for strings in the row
    for value in row:
        if isinstance(value, str):
            print("String found:", value)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr_mat = df.corr()
#print('pearson correlation = \n', corr_mat, sep='')
```

```python
plt.figure(figsize=(13,9))
ax = sns.heatmap(data=corr_mat, annot=True,
            cmap='GnBu', annot_kws={"weight": "bold"})
plt.title('Correlation Matrix for Auto MPG dataset')
plt.show()
```
